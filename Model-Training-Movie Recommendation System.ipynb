```python
import pandas as pd            # importing required libraries
import numpy as np
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')          # reading the movies.csv file
credits = pd.read_csv('tmdb_5000_credits.csv')        # reading the credits.csv file
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')          # merge both 'movie' & 'credits' datasets through title collum
```

```python
movies.head(1)
```

```python
#######             Colums we need
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]    #fetsh only the listed colums in movies 
```

```python
movies.info()
```

```python
movies.head(1)
```

```python
movies.isnull().sum()         #checking if the data has any null values
                              # we see that overview has 3 null values
```

```python
movies.dropna(inplace=True)     #we handle the null values by dropping the colums
```

```python
movies.isnull().sum()               #now our null value problem is solved
```

```python
movies.duplicated().sum()            #checking if the data has any duplicate values 
                                     # we see that which is 0
```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# we need to extract ['action','adventure','fantasy','scifi'] keywords 
```

```python
def convert(obj):      # we created a convert function to convert above string data into list of keywords we needed
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)       # we apply the convert function into genres colume and update data
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)   # we do the same thing with keyword colume as we did with genres colume
```

```python
movies.head(1)
```

```python
def convert3(obj):         # we created a new function to extract first 3 cast
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):       # we created fetch_director function to extract director name from data
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)      # we apply the function and update the data
```

```python
movies.head(1)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())      # we convert the overview colume date into list
```

```python
movies.head(1)
```

```python
# now we need to make a transformation 
# we need to convert 'sam worthington' into 'samworthington'
# reason: for machine 'sam wothington' are two different person But in reality they are same person
# we need to apply this not only in crew colume but also in genres & keywords colume also.
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
# now we need to create a new colume 'tags' and we concatenate overvies,genres,keywords,cast,crew columes in it.

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
# now we created a new dataframe in which we include movies_id, title , tags colume

new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
# now we convert the tags colume data from list to string

new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
# we convert all the tags data into lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# now we perform the text vectorisation
# we calculate the similarities between the tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# now we get a problem you can see above there are many words with same meaning
# like 'adult' & 'adults' or 'love' & 'loved'
# for solving this we apply steamming on the 'tags' colume
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
